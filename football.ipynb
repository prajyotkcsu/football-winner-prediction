```python
# Becasue ML can work only on numeric data and not on object dtypes, so the only choose we have is either work with only numeric data as predictors or somehow standardizing objects as numbers. In iur case we can convert venue as category(0,1) opponents as numbers 0 to 15.
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python


df['venue_code']=df['venue'].astype('category').cat.codes
```

```python
df
```

```python
df['opp_code']=df['opponent'].astype('category').cat.codes
```

```python
df
```

```python
# because some teams might play better at a specific hour of the day, we will convert hour into  only number ef 16:30 to 16. this in a way puts hours in catefgory of fixed size
```

```python
df['hour']=df['time'].str.replace(':.+','',regex=True).astype('int')
```

```python
df
```

```python
df['day_code']=df['day'].astype('category').cat.codes
```

```python
df
```

```python
df
```

```python
#since we are predecting if the team Win or Loss, we consider draw as not win so we have just two categorize (0,1`)
```

```python
df['target']=(df['result'] == 'W').astype("int")
```

```python
df
```

```python
#ready to do machine learning
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#random forrest, n_estimators is number of those many decision trees being trained on simultaneously with each tree having different set of parameters
#higher the number is more time it will take to run the model.
rf=RandomForestClassifier(n_estimators=50, min_samples_split=10,random_state=1)
```

```python
train=df[df['date']< '2022-01-01']
```

```python
test=df[df['date']> '2022-01-01']
```

```python
predictors=['venue_code', 'opp_code', 'day_code']
```

```python
rf.fit(train[predictors],train['target'])
```

```python
preds=rf.predict(test[predictors])
```

```python
# accuracy_score tells How many times did you predict the team win vs How many times did it actually win.
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(test['target'], preds)
```

```python
acc
```
